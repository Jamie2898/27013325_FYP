```python
import matplotlib.pyplot as plt
```

```python
# ===================================
# Baseline
# Male Binary Classification
# ===================================


x1=['SVM','LR','DT','RF']
y1=[69,76,85,73]
x2=['SVM','LR','DT','RF']
y2=[81,85,83,90]
x3=['SVM','LR','DT','RF']
y3=[69,69,67,77]
plt.figure(figsize=(9.6, 7.2))
l1=plt.plot(x1,y1,'r--',label='CN-MCI')
l2=plt.plot(x2,y2,'g--',label='CN-AD')
l3=plt.plot(x3,y3,'b--',label='MCI-AD')
plt.plot(x1,y1,'ro-',x2,y2,'g+-',x3,y3,'b^-')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Baseline - Male Binary Classification', fontsize=15)
plt.xlabel('Classifier', fontsize=15)
plt.ylabel('Accuracy', fontsize=15)
plt.legend(loc=2, prop={'size': 15})
plt.show()
```
